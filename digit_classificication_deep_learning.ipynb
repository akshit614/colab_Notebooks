```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
```

```python
(X_train, y_train), (X_test, y_test) = keras.datasets.mnist.load_data()
```

```python
len(X_train)
```

```python
len(y_test)
```

```python
plt.matshow(X_train[0])
```

```python
X_train[0]
```

```python
X_train.shape
```

```python
X_train = X_train / 255
X_test = X_test / 255
```

```python
X_train[0]
```

```python
X_train_falttened = X_train.reshape(len(X_train) , 28*28)
X_test_falttened = X_test.reshape(len(X_test) , 28*28)
X_train_falttened.shape
```

```python
X_train_falttened[0]
```

```python
model = keras.Sequential([
    keras.layers.Dense(10, input_shape = (784,), activation = 'sigmoid')
])

model.compile(
    optimizer = 'adam',
    loss = 'sparse_categorical_crossentropy',
    metrics = ['accuracy']
)

model.fit(X_train_falttened, y_train, epochs = 10)
```

```python
model.evaluate(X_test_falttened, y_test)
```

```python
y_predict = model.predict(X_test_falttened)
y_predict[0]
```

```python
np.argmax(y_predict[0])
```

```python
y_pred = [np.argmax(i) for i in y_predict]
```

```python
cm = tf.math.confusion_matrix(y_test, y_pred)
cm
```

```python
import seaborn as sns
plt.figure(figsize = (10,5))
sns.heatmap(cm, annot= True, fmt = 'd')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

```python
model2 = keras.Sequential([
    keras.layers.Dense(100, input_shape = (784,), activation = 'relu'),
    keras.layers.Dense(10, activation = 'sigmoid')
])

model2.compile(
    optimizer = 'adam',
    loss = 'sparse_categorical_crossentropy',
    metrics = ['accuracy']
)

model2.fit(X_train_falttened, y_train, epochs = 5)
```

```python
model2.evaluate(X_test_falttened, y_test)
```

```python
y_pred = model2.predict(X_test_falttened)
y_pred[0]
```

```python
np.argmax(y_pred[0])
```

```python
y_predictt = [np.argmax(i) for i in y_pred]
```

```python
cm2 = tf.math.confusion_matrix(y_test, y_predictt)
cm2
```

```python
plt.figure(figsize = (10,5))
sns.heatmap(cm2, annot= True, fmt = 'd')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

```python
model2 = keras.Sequential([
    keras.layers.Flatten(input_shape=(28,28),),
    keras.layers.Dense(100, input_shape = (784,), activation = 'relu'),
    keras.layers.Dense(10, activation = 'sigmoid')
])

model2.compile(
    optimizer = 'adam',
    loss = 'sparse_categorical_crossentropy',
    metrics = ['accuracy']
)

model2.fit(X_train, y_train, epochs = 10)
```

```python
model2.evaluate(X_test,y_test)
```

```python
y_pred = model2.predict(X_test)
np.argmax(y_pred[0])
```

```python
y_predictt = [np.argmax(i) for i in y_pred]
cm3 = tf.math.confusion_matrix(y_test, y_predictt)
cm3
```

```python
plt.figure(figsize = (10,7))
sns.heatmap(cm3, annot= True, fmt = 'd')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
